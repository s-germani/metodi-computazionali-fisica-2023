### Operazioni con Array 

`numpy` permette di effettuare operazioni matematiche sugli `array` che si applicano elemento per elemento (differente da operzioni fra vettori). 
Gli `array` devono necessariamente essere della stessa dimensione.

Sono permesse anche operazioni fra `array` e scalarim che si applicano ad ognui elemento.
 

```python
# dfinisci array di partenza
aa1 = np.arange(10)
aa2 = np.ones(10)

print('aa1', aa1)
print('aa2', aa2)

#somma con scalare
aa3 = aa1 +0.25
print('somma con scalare', aa3)

# somma tra vettori
aasum = aa1+aa2
print('somma', aasum)
```

```python
# differenza fra vettori
aadiff = aa1-aa2
print('diff', aadiff)
```

```python
# prodotto con scalare 

aadouble = aa1*2

print('prodotto con scalare', aadouble)
```

```python
# prodotto fra vettori 

aa4 = np.logspace(-3, 3, 10)

aaproduct = aa1*aa4

print('prodotto fra vettori', aaproduct)
```

```python
# operazioni fra array di diverse dimensioni non sono permesse
aa2 = np.ones(15)

aa1+aa2[:10]
```

```python
# potenze

ab = np.arange(1,5)
asq = ab**2
asq
```

```python
atrd = ab**3
atrd
```

```python

np.power(ab, 2)

np.power(ab, ab)
```

```python
# seno
np.sin(ab)
```

```python
#logaritmo
np.log(ab)
```

```python
# ordinare gli elementi
myarray = np.array([  22,33,11, 99, 55, 77, 44, 1])

np.sort(myarray)
```

### Maschere

Per effettuare operazioni o utilizzare solo elementi degli `array` che soddsfano determinate condizioni si possono utlizzare le maschere.

```python
am = np.array([ 10, 20, 11, 1, 22, 2, 7, 77, 101, 274, 888, 55, 4])

# definizione maschera di slezione
amask = am > 10

amask
```

```python
~amask
```

```python
# utilizzo maschera
print('array originale           ', am)
print('array con maschera        ', am[amask])
print('array con maschera inversa', am[~amask])
```

```python
len(am)
```

### Test di Velocità

```python
import time
```

```python

avtest1 = np.arange(1e3)

start_time1 = time.time()
sumnp = avtest1.sum()
print("--- numpy {:} {:2.6f} seconds ---".format( sumnp, time.time() - start_time1))


start_time2 = time.time()
sum1 = 0
for ae in avtest1:
    sum1 = sum1 + ae

print("--- somma {:} {:2.6f} seconds ---".format(sum1, time.time() - start_time2))
```

```python
avtest1 = np.arange(1e6)

start_time1 = time.time()
sumnp = avtest1.sum()
print("--- numpy {:} {:2.6f} seconds ---".format( sumnp, time.time() - start_time1))


start_time2 = time.time()
sum1 = 0
for ae in avtest1:
    sum1 = sum1 + ae

print("--- somma {:} {:2.6f} seconds ---".format(sum1, time.time() - start_time2))

#print( 1e6*(1e6-1)/2)
```

Le funzionalità di `pandas` sono costruite attorno ai ***DataFrame*** che corrispondono agli oggetti  utilizzati per immagazzinare e manipolare i dati. 

```python
import pandas as pd
```

```python
# Creo array per colonne DataFrame
cc1 = np.arange(10, 20)
cc2 = np.linspace(-2,2, 10)
cc3 = np.logspace(1, 3, 10)
```

## Matplotlib

Principale modulo `Python` per la produzione di grafici.

```python
import matplotlib.pyplot as plt
```

```python
# plot
ax = np.arange(100)

ay = 2+ax**2.5

plt.plot(ax,ay)
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
# assi logaritmici 

alx = np.logspace(-1, 5)
aly = alx+np.power(alx,2)
plt.plot(alx,aly, color='orange')
plt.xlabel('X')
plt.ylabel('Y')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
# plot
ax1 = np.arange(10)

ay1 = 5+10*ax1**0.5
ay2 = 2+ax1**2

plt.plot(ax1, ay1, 'o-', color='limegreen', label='Y1' )
plt.plot(ax1, ay2, 'o',  color='red' ,      label='Y2' )

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
#subplots

fig,ax = plt.subplots(1,2, figsize=(12,6) )
print('ax shape', ax.shape)

ax[0].plot(ax1, ay1, 's-', color='limegreen')
ax[1].plot(ax1, ay2, '*',  color='red'  )

ax[0].set_title('Grafico Y1', fontsize=15, color='limegreen')
ax[1].set_title('Grafico Y2', fontsize=15, color='red')

ax[0].set_xlabel('X1')
ax[0].set_ylabel('Y1')

ax[1].set_xlabel('X2')
ax[1].set_ylabel('Y2')

ax[0].tick_params(axis='x', labelsize=14)
ax[0].tick_params(axis='y', labelsize=14)
ax[0].set_xlim(-1, 10)
ax[0].set_ylim(0, 40)
ax[0].grid(True)

plt.show()
```

```python
# errorbar
ey1 = 1 + 0.1 * ay1

plt.errorbar(ax1, ay1, yerr=ey1, fmt='o' )

plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# scatter 

xs = np.linspace(-10, 30, 11)

ysb = np.array([22, 33, 11, 66, 44, 12, 19, 55, 88, 31, 50])
ysr = np.array([12, 23, 21, 56, 54, 62, 79, 85, 18, 71, 40])

plt.scatter( xs, ysb, color='royalblue', s=32, label='piccoli')
plt.scatter( xs, ysr, color='tomato',    s=68, label='grandi')
plt.xlabel('X', fontsize=16)
plt.ylabel('Y', fontsize=16)
plt.ylim(0, 100)
plt.legend(fontsize=14)
plt.show()
```
