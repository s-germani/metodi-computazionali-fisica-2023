### Istruzioni condizionali (if-elif-else)

```python
avc = 10

if avc > 20:
    print(avc, '> 20')
    
```

```python
if avc < 20:
    print(avc, '< 20')
```

```python

bvc = 20
if bvc > 20:
    print('-->', bvc, '> 20')
elif bvc < 20:
    print('-->', bvc, '< 20')
else:
    print('-->', bvc, '= 20')
    
```

```python
for k in range(10):
    if k >=1 and k < 5: 
        print(k, '<<<')
    elif k == 5:
        print(k, '===')
    elif k> 5 and k < 9:
        print(k, '>>>')
    else:
        print(k, '***')
```

### Formattazione Stringhe

```python
si = 10
sf = 3.14159265359 

pippo    = 'Pippo'
pluto    = 'Pluto'
paperino = 'Paperino'
```

```python
# Formattazione generica con tipo di dato non specificato
mysg = 'Esempio di formattazione generica -->{:} - {:}  --  {:}'.format(si, pippo, sf)
print(mysg)
```

```python
# Formattazione con interi base
mysi1 = 'Esempio di formattazione con interi {:d}'.format(si)
print(mysi1)
```

```python
# Formattazione con interi e spazio fisso riempiti con zeri
mysi2 = 'Secondo esempio di formattazione con interi {:04d}'.format(si)
print(mysi2)
```

```python
# Formattazione con interi e spazio fisso
mysi3 = 'Terzo esempio di formattazione con interi{:20d}'.format(si)
print(mysi3)
```

```python
# Formattazione con float
mysf1 = 'Esempio di formattazione con float {:f}'.format(sf)
print(mysf1)
```

```python
# Formattazione con float e specifica su cifre significative 
mysf2 = 'Secondo esempio di formattazione con float {:.2f}'.format(sf)
print(mysf2)
```

```python
# Formattazione con float e specifica su cifre significative 
mysf3 = 'Terzo esempio di formattazione con float {:.10f}'.format(sf)
print(mysf3)
```

```python
# Formattazione con stringa
mysf1 = 'Esempio di formattazione con stringhe {:s}'.format(pippo)
print(mysf1)
```

```python
# Formattazione con stringa
mysf2 = 'Esempio di formattazione con stringhe {:s}, {:s}'.format(pippo, pluto.upper())
print(mysf2)
```

```python
# Formattazione con numerazione campi 
mysf1 = 'Esempio di formattazione  {0} -  {1} - {0} - {2}'.format(pippo, pluto, paperino)
print(mysf1)
```

```python
ll = [pippo, pluto, paperino]

for s in ll:
    print(s, len(s))
```

```python
# Allineamento a sinistra
for s in ll:
    print('{:<10} {:02d}'.format(s, len(s)))
```

```python
# Allineamento a destra
for s in ll:
    print('{:>10} {:02d}'.format(s, len(s)))
```

```python
# Allineamento al centro
for s in ll:
    print('{: ^10} {:02d}'.format(s, len(s)))
```

```python
# Allineamento al centro
for s in ll:
    print('{:-^10} {:02d}'.format(s, len(s)))
```

## Funzioni

```python
# Definizione funzione fprova
def fprova():
    print('Funzione di Prova')
```

```python
# Chiamata funzione fprova
fprova()
```

```python
# Definizione funzione fprova_input che accetta un valore in ingresso
def fprova_input(vi):
    print('Funzione di Prova con input: {:}'.format(vi))
```

```python
# Chiamate funzione fprova_input
fprova_input(11)
fprova_input('CIAO')
```

```python
# Definizione funzione che restituisce valore (somma)
def somma(a,b):
    return a+b
```

```python
# Chiamata funzione somma
mysum = somma(2, 3)
```

```python
print(mysum)
```

```python
# Definizione funzione che restituisce pi valori
def powe(b, e):
    v1 = b**e
    v2 = e *b**e
    v3 = e**2 * b**e
    return v1,v2, v3
```

```python
a,b,c = powe(3, 2)
```

```python
print(a)
print(b)
print(c)
```

## List Comprehension

```python
mylist = [2, 4, 8, 11, 32]
```

```python
new_list1 = [ v+3 for v in mylist ]
print(new_list1)
```

```python
new_list2 = [ v-3 for v in mylist if v-3>0]
print(new_list2)
```
